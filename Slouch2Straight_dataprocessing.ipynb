```python
from google.colab import drive
import scipy.io as sio
#mount your drive so google colab can access files in it
drive.mount('/content/drive')
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#load that data set stored in test_data_straight1.mat
dir = "/content/drive/MyDrive/UCLA_ACADEMIC/M119/Align/DATA/"

data1 = pd.read_csv(dir + 'test_data_slouchtostr1.csv')
data2 = pd.read_csv(dir + 'test_data_slouchtostr2.csv')
data3 = pd.read_csv(dir + 'test_data_slouchtostr3.csv')
data4 = pd.read_csv(dir + 'test_data_slouchtostr4.csv')
data5 = pd.read_csv(dir + 'test_data_slouchtostr5.csv')
```

```python
print(data1.keys())

length1 = data1.shape[0]
length2 = data2.shape[0]
length3 = data3.shape[0]
length4 = data4.shape[0]
length5 = data5.shape[0]

t1 = np.linspace(0,1,length1)
t2 = np.linspace(0,1,length2)
t3 = np.linspace(0,1,length3)
t4 = np.linspace(0,1,length4)
t5 = np.linspace(0,1,length5)



Ax1 = data1['Ax']
Ay1 = data1[' Ay']
Az1 = data1[' Az']
Gx1 = data1[' Gx']
Gy1 = data1[' Gy']
Gz1 = data1[' Gz']

Ax2, Ay2, Az2, Gx2, Gy2, Gz2 = data2['Ax'], data2[' Ay'], data2[' Az'], data2[' Gx'], data2[' Gy'], data2[' Gz']
Ax3, Ay3, Az3, Gx3, Gy3, Gz3 = data3['Ax'], data3[' Ay'], data3[' Az'], data3[' Gx'], data3[' Gy'], data3[' Gz']
Ax4, Ay4, Az4, Gx4, Gy4, Gz4 = data4['Ax'], data4[' Ay'], data4[' Az'], data4[' Gx'], data4[' Gy'], data4[' Gz']
Ax5, Ay5, Az5, Gx5, Gy5, Gz5 = data5['Ax'], data5[' Ay'], data5[' Az'], data5[' Gx'], data5[' Gy'], data5[' Gz']

```

```python
#plot 
fig, ax = plt.subplots(2,3, figsize=(25,20))



ax[0,0].plot(t1, Ax1)
ax[0,0].set_title("Ax")
ax[0,0].grid()

#ax[0,0].set_ylim([-20,20]) 

ax[0,1].plot(t1, Ay1)
ax[0,1].set_title("Ay")
ax[0,1].grid()

ax[0,2].plot(t1, Ay1)
ax[0,2].set_title("Ay")
ax[0,2].grid()

ax[1,0].plot(t1, Gx1)
ax[1,0].set_title("Gx")
ax[1,0].grid()
ax[1,1].plot(t1, Gy1)
ax[1,1].set_title("Gy")
ax[1,1].grid()
fig.tight_layout(pad=4.0)
ax[1,2].plot(t1, Gz1)
ax[1,2].set_title("Gz")
ax[1,2].grid()

for i in range(3):
    ax[1,i].set_ylim([-10,30]) 

for j in range(3):
    ax[0,j].set_ylim([-1,1]) 

plt.show()
```
